### Colorful Floral Pattern

This code generates a mesmerizing floral pattern using `ColabTurtle`. It sets a dark background and cycles through a list of colors. The pattern is created by repeatedly drawing lines of increasing length and turning at precise angles, which leads to a complex and beautiful symmetrical design.

```python
from ColabTurtle.Turtle import *

initializeTurtle(initial_speed=13, initial_window_size=(700, 700))
bgcolor('darkslategrey') # A calming dark background
speed(10) # Set a good drawing speed

colors = ['#FF0000', '#FF7F00', '#FFFF00', '#00FF00', '#0000FF', '#4B0082', '#9400D3'] # Rainbow colors
width(2) # Slightly thicker lines for visibility

penup()
setposition(350, 350) # Start in the center of the window
pendown()

for i in range(360): # Loop many times to create a dense pattern
    pencolor(colors[i % len(colors)]) # Cycle through the rainbow colors
    forward(i * 0.8) # Increase line length gradually for a spiral effect
    left(135) # A specific turn angle to create petal-like shapes (e.g., 135, 144, 150 can make interesting patterns)
```

### Koncentrick rotujc tverce

Tento kd vytv dynamick vzor koncentrickch tverc, kter se postupn zvtuj a mrn rotuj. Kad tverec je nakreslen jinou barvou z definovanho seznamu. Nastaven `darkblue` pozad pomh kontrastu s jasnmi barvami tverc, vytv zajmav vizuln efekt, kter se li od spirl a hvzd.

```python
from ColabTurtle.Turtle import *

initializeTurtle(initial_speed=13, initial_window_size=(600, 600))
bgcolor('darkblue') # Tmav modr pozad pro vrazn kontrast
speed(10) # Maximln rychlost kreslen

colors = ['cyan', 'lime', 'yellow', 'orange', 'red', 'magenta'] # Sada jasnch barev
width(2) # Stedn tlouka ry

penup()
setposition(300, 300) # Zanme v pesnm stedu okna
pendown()

side_length = 10 # Poten dlka strany tverce
angle_increment = 5 # hel rotace pro kad dal tverec
num_squares = 60 # Poet tverc, kter se nakresl

for i in range(num_squares):
    pencolor(colors[i % len(colors)]) # Cyklick zmna barvy pro kad tverec
    # Nakreslete jeden tverec
    for _ in range(4):
        forward(side_length)
        left(90)
    left(angle_increment) # Mrn otote elvu pro dal tverec
    side_length += 5      # Zvtete dlku strany pro dal tverec, aby se rozioval ven
```

### Dynamick organick spirla

Tento kd generuje uniktn organickou spirlu. Na tmavm pozad se elva pohybuje vped a jej dlka kroku se postupn zvtuje. Klovm prvkem je, e hel otoen se s kadm krokem mrn mn, co vytv nepravidelnou a dynamickou spirlu namsto geometricky pesnho tvaru. Barvy se cyklicky mn, co pidv vizuln komplexnost.

```python
from ColabTurtle.Turtle import *

initializeTurtle(initial_speed=13, initial_window_size=(700, 700))
bgcolor('darkslateblue') # Tmav modrofialov pozad
speed(10) # Maximln rychlost kreslen

colors = ['gold', 'red', 'lightgreen', 'deepskyblue', 'magenta', 'orange'] # Nov sada ivch barev
width(2)

penup()
setposition(350, 350) # Zanme uprosted okna
pendown()

for i in range(300): # Hodn iterac pro hustotu vzoru
    pencolor(colors[i % len(colors)]) # Cyklick zmna barev
    forward(i * 0.7) # Dlka ry se postupn zvyuje
    left(i / 10 + 90) # hel se mn s kadou iterac, vytv dynamickou spirlu
```

### Kompozice: Dynamick spirla a rotujc tverce

Tento kd vytv komplexn obrazec, kter kombinuje nkolik rznch prvk na jednom pltn. Uprosted je dynamick, organicky se vyvjejc spirla s mncm se hlem oten a barvami. V kadm ze ty roh jsou pak men sady koncentrickch rotujcch tverc, kad s vlastn barevnou paletou. Pozad je tmav pro zvraznn vech ivch barev.

```python
from ColabTurtle.Turtle import *
import random

initializeTurtle(initial_speed=13, initial_window_size=(800, 800)) # Vt pltno
bgcolor('darkslategrey') # Tmav, ale ne ern pozad
speed(10)

# --- Funkce pro kreslen dynamick spirly ---
def draw_dynamic_spiral(x, y, num_iterations, base_length, angle_modifier, colors):
    penup()
    setposition(x, y)
    pendown()
    for i in range(num_iterations):
        pencolor(colors[i % len(colors)])
        forward(base_length + i * 0.7)
        left(i / angle_modifier + 90)

# --- Funkce pro kreslen koncentrickch rotujcch tverc ---
def draw_rotating_squares(x, y, num_squares, start_side_length, angle_increment, size_increment, colors):
    penup()
    setposition(x, y)
    pendown()
    current_side_length = start_side_length
    for i in range(num_squares):
        pencolor(colors[i % len(colors)])
        for _ in range(4):
            forward(current_side_length)
            left(90)
        left(angle_increment)
        current_side_length += size_increment

# --- Funkce pro kreslen paprsk (sunburst) ---
def draw_sunburst(x, y, num_rays, ray_length, colors):
    penup()
    setposition(x, y)
    pendown()
    setheading(0) # Zanme s elvou smujc na vchod
    angle_step = 360 / num_rays
    for i in range(num_rays):
        pencolor(colors[i % len(colors)])
        forward(ray_length)
        backward(ray_length) # Vrtme se do stedu
        left(angle_step)

# --- Nov funkce pro kreslen radilnch zblesk v kvadrantech ---
def draw_radial_burst(x, y, num_lines, line_length, colors):
    penup()
    setposition(x, y)
    pendown()
    original_heading = heading() # Ulome pvodn smr
    angle_step = 360 / num_lines
    for i in range(num_lines):
        pencolor(colors[i % len(colors)])
        forward(line_length)
        backward(line_length)
        left(angle_step)
    setheading(original_heading) # Obnovme pvodn smr

# --- Nov funkce pro spojen bod rami do tverce ---
def draw_connecting_square(point1, point2, point3, point4, color, line_width):
    penup()
    setposition(point1[0], point1[1])
    pendown()
    pencolor(color)
    width(line_width)
    setposition(point2[0], point2[1])
    setposition(point3[0], point3[1])
    setposition(point4[0], point4[1])
    setposition(point1[0], point1[1]) # Zpt na zatek pro uzaven tverce

# --- Nov funkce pro kreslen roziujcch se hvzd (pro konkrtn efekty) ---
def draw_expanding_stars(x, y, num_stars, num_points, start_size, size_increment, angle_increment, colors):
    penup()
    setposition(x, y)
    pendown()
    current_size = start_size
    for i in range(num_stars):
        pencolor(colors[i % len(colors)])
        for _ in range(num_points):
            forward(current_size)
            right(360 / num_points * 2) # Pro 5cpou hvzdu je to 144 stup (360/5 * 2)
            forward(current_size)
            left(360 / num_points) # Aby se vrtil do vchozho smru pro dal rameno
        left(angle_increment) # Rotace cel hvzdy
        current_size += size_increment # Zvten velikosti

# --- Nov funkce pro kreslen jedn statick hvzdy ---
def draw_static_single_star(x, y, size, num_points, color):
    penup()
    setposition(x, y)
    pencolor(color)
    width(1)
    pendown()
    for _ in range(num_points):
        forward(size)
        right(360 / num_points * 2)
        forward(size)
        left(360 / num_points)
    penup()

# --- Nov funkce pro kreslen mal, hust spirly ---
def draw_small_spiral_corner(x, y, num_iterations, initial_length, angle, colors, pen_thickness):
    penup()
    setposition(x, y)
    pendown()
    width(pen_thickness)
    for i in range(num_iterations):
        pencolor(colors[i % len(colors)])
        forward(initial_length + i * 0.5)
        left(angle)

# --- Nov funkce pro kreslen malch polygonovch 'kvtin' ---
def draw_small_polygon_flower(x, y, size, num_sides, rotation_angle, colors, line_width):
    penup()
    setposition(x, y)
    pendown()
    pencolor(random.choice(colors))
    width(line_width)
    setheading(random.randint(0, 359)) # Nhodn poten smr
    left(rotation_angle) # Poten rotace
    for _ in range(num_sides):
        forward(size)
        left(360 / num_sides)
    penup()

# --- Hlavn kompozice ---

# 1. Centrln dynamick spirla
spiral_colors = ['gold', 'red', 'lightgreen', 'deepskyblue', 'magenta', 'orange']
draw_dynamic_spiral(400, 400, 200, 0, 10, spiral_colors)

# 2. Rotujc tverce v rozch
square_colors_1 = ['cyan', 'skyblue', 'blue', 'darkblue']
draw_rotating_squares(150, 150, 20, 5, 10, 3, square_colors_1) # Lev horn

square_colors_2 = ['lime', 'green', 'darkgreen', 'forestgreen']
draw_rotating_squares(650, 150, 20, 5, 10, 3, square_colors_2) # Prav horn

square_colors_3 = ['yellow', 'orange', 'darkorange', 'chocolate']
draw_rotating_squares(150, 650, 20, 5, 10, 3, square_colors_3) # Lev doln

square_colors_4 = ['magenta', 'purple', 'darkviolet', 'indigo']
draw_rotating_squares(650, 650, 20, 5, 10, 3, square_colors_4) # Prav doln

# 3. Pidme Sunburst efekty do stedu kad strany
sunburst_colors = ['white', 'silver', 'grey']
draw_sunburst(400, 100, 15, 50, sunburst_colors) # Horn sted
draw_sunburst(400, 700, 15, 50, sunburst_colors) # Doln sted
draw_sunburst(100, 400, 15, 50, sunburst_colors) # Lev sted
draw_sunburst(700, 400, 15, 50, sunburst_colors) # Prav sted

# 4. Pidme radiln zblesky do kvadrant
radial_colors = ['lightblue', 'lightcoral', 'lightgreen', 'lavender']
draw_radial_burst(275, 275, 12, 40, radial_colors) # Lev horn kvadrant
draw_radial_burst(525, 275, 12, 40, radial_colors) # Prav horn kvadrant
draw_radial_burst(275, 525, 12, 40, radial_colors) # Lev doln kvadrant
draw_radial_burst(525, 525, 12, 40, radial_colors) # Prav doln kvadrant

# 5. Spojme stedy radilnch zblesk rami do tverce
connect_color = 'white'
connect_width = 2

point_lh = (275, 275) # Lev horn
point_ph = (525, 275) # Prav horn
point_ld = (275, 525) # Lev doln
point_pd = (525, 525) # Prav doln

draw_connecting_square(point_lh, point_ph, point_pd, point_ld, connect_color, connect_width)

# 6. Pidme roziujc se hvzdy do zbvajcch roh (pvodn pouit)
star_colors_expanding = ['gold', 'red', 'lightgreen', 'blue', 'magenta']
draw_expanding_stars(200, 200, 10, 5, 10, 5, 15, star_colors_expanding) # Mezi levm hornm tvercem a sunburstem
draw_expanding_stars(600, 200, 10, 5, 10, 5, 15, star_colors_expanding) # Mezi pravm hornm tvercem a sunburstem

# 7. Pidme mal statick hvzdiky po celm obrzku k vyplnn przdnch mst
small_star_colors = ['white', 'yellow', 'lightblue', 'pink']
num_small_stars = 100
for _ in range(num_small_stars):
    star_x = random.randint(50, 750)
    star_y = random.randint(50, 750)
    star_size = random.randint(5, 15)
    star_points = random.choice([4, 5, 6]) # Rzn poty cp
    star_color = random.choice(small_star_colors)
    draw_static_single_star(star_x, star_y, star_size, star_points, star_color)

# 8. Pidme mal hust spirly do plnch roh pltna
corner_spiral_colors = ['white', 'silver', 'cyan', 'magenta']
draw_small_spiral_corner(50, 50, 50, 1, 89, corner_spiral_colors, 1)    # Lev horn
draw_small_spiral_corner(750, 50, 50, 1, 89, corner_spiral_colors, 1)  # Prav horn
draw_small_spiral_corner(50, 750, 50, 1, 89, corner_spiral_colors, 1)  # Lev doln
draw_small_spiral_corner(750, 750, 50, 1, 89, corner_spiral_colors, 1) # Prav doln

# 9. Pidme mal polygonov 'kvtiny' po celm pltn pro vyplnn mezer
flower_colors = ['#FFC0CB', '#DA70D6', '#BA55D3', '#9370DB', '#8A2BE2', '#4B0082'] # Odstny rov a fialov
num_flowers = 80
for _ in range(num_flowers):
    flower_x = random.randint(20, 780)
    flower_y = random.randint(20, 780)
    flower_size = random.randint(5, 20)
    flower_sides = random.choice([3, 4, 5, 6]) # Trojhelnky, tverce, ptihelnky, estihelnky
    flower_rotation = random.randint(0, 90)
    draw_small_polygon_flower(flower_x, flower_y, flower_size, flower_sides, flower_rotation, flower_colors, 1)
```
